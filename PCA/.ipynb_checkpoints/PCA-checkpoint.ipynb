```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Data
df = pd.read_csv('../Data/train.csv')
```

```python
df.head(5)
```

```python
y = df.loc[:,['SalePrice']].values
x = df.loc[:, ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt',\
               'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', \
               'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',\
               'Fireplaces', 'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF']]
x = x.dropna(axis='columns')
```

```python
pca = PCA(n_components = 5)

principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
```

```python
principalDf
```

```python
finalDf = pd.concat([principalDf, df[['SalePrice']]], axis = 1)
```

```python
ComponentsMatrix = pd.DataFrame(pca.components_,columns=x.columns,index = ['PC1','PC2','PC3','PC4','PC5'])
```

```python
ComponentsMatrix.columns
```

```python
pc1_exp = pd.DataFrame({'Componentes':ComponentsMatrix.columns, 'Importance':ComponentsMatrix.iloc[0]})

ax1 = pc1_exp.plot.bar(x='Componentes', y='Importance', rot=90)
```

```python
data = pd.DataFrame({'Componentes':['PC1', 'PC2', 'PC3', 'PC4', 'PC5'], 'porcentaje':pca.explained_variance_ratio_})

ax = data.plot.bar(title='Component impact on model',x='Componentes', y='porcentaje', rot=0)
```
