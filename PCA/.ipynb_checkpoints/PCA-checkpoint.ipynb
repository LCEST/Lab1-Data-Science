### Variables dependientes e independientes

```python
# Data
df = pd.read_csv('../Data/train.csv')
```

```python
y = df.loc[:,['SalePrice']].values

x = df.loc[:, ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt',\
               'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',\
               '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', \
               'BsmtHalfBath', 'FullBath', 'HalfBath','Fireplaces', 'GarageYrBlt', \
               'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'YrSold']]

temp_x = df.loc[:, ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt',\
               'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',\
               '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', \
               'BsmtHalfBath', 'FullBath', 'HalfBath','Fireplaces', 'GarageYrBlt', \
               'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'YrSold', 'SalePrice']]

x = x.dropna(axis='rows')
temp_x = temp_x.dropna(axis='rows')
```

### Matriz de Correlaci√≥n

### Bartlett's Test

H0: |R| = 1

H1: |R| != 1

Criterio de la prueba p < Œ±. Como se quiere un nivel de significancia del 95% se estar√° utilizando 0.05 como Œ±. 

```python
stat, p = bartlett(temp_x.iloc[:,0], temp_x.iloc[:,1], temp_x.iloc[:,2], temp_x.iloc[:,3], temp_x.iloc[:,4], temp_x.iloc[:,5], temp_x.iloc[:,6],
                   temp_x.iloc[:,7], temp_x.iloc[:,8], temp_x.iloc[:,9], temp_x.iloc[:,10], temp_x.iloc[:,11], temp_x.iloc[:,12], temp_x.iloc[:,13],
                   temp_x.iloc[:,14], temp_x.iloc[:,15], temp_x.iloc[:,16], temp_x.iloc[:,17], temp_x.iloc[:,18], temp_x.iloc[:,19], temp_x.iloc[:,20],
                   temp_x.iloc[:,21], temp_x.iloc[:,22], temp_x.iloc[:,23], temp_x.iloc[:,24], temp_x.iloc[:,25], temp_x.iloc[:,26], temp_x.iloc[:,27])
p
```

Ya que p < 0.05, se rechaza H0 y se acepta H1.

### Divisi√≥n en Componentes

```python
x = pd.DataFrame(StandardScaler().fit_transform(x), columns=x.columns)
```

```python
pca = PCA(n_components = 7)

principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7'])
```

```python
principalDf
```

```python
finalDf = pd.concat([principalDf, df[['SalePrice']]], axis = 1)
```

```python
ComponentsMatrix = pd.DataFrame(pca.components_,columns=x.columns,index = ['PC1','PC2','PC3','PC4','PC5','PCA6','PCA7'])
```

```python
ComponentsMatrix.columns
```

### Variables v. Componente
